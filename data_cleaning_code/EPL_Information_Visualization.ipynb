```python
# Imports
import pandas as pd
```

```python
# EPL League Standings Dataset
epl_league_standings_df = pd.read_csv('EPL_League_Standings.csv')
epl_league_standings_df = epl_league_standings_df[['year','Team','A','D','F','GD','L','P','Pos','Pts','W']]
epl_league_standings_df.columns = ['epl_season','epl_team','goals_conceded','matches_drawn','goals_scored','goal_difference','matches_lost','matches_played','season_position','season_points','matches_won'] 
epl_season_filter = ['2006/2007','2007/2008','2008/2009','2009/2010','2010/2011','2011/2012','2012/2013','2013/2014','2014/2015','2016/2017']
epl_league_standings_df = epl_league_standings_df[epl_league_standings_df['epl_season'].isin(epl_season_filter)]
epl_league_standings_df
```

```python
# Adding the data of the missing season 2015-2016 using another Dataset
epl_league_standings_2015_2016_df = pd.read_csv('epl20152016.csv')
epl_league_standings_2015_2016_df.columns = ['season_position','epl_team','matches_played','matches_won','matches_drawn','matches_lost','goals_scored','goals_conceded','goal_difference','season_points']
epl_league_standings_2015_2016_df['epl_season'] = '2015/2016'
epl_league_standings_2015_2016_df
```

```python
# Checking if there are NaN values in epl_league_standings_df 
epl_league_standings_df.isnull().values.any()
```

```python
# Checking if there are NaN values in epl_league_standings_2015_2016_df 
epl_league_standings_2015_2016_df.isnull().values.any()
```

```python
# Merging both the datasets
epl_league_standings_df = pd.concat([epl_league_standings_df,epl_league_standings_2015_2016_df],sort=True).reset_index(drop=True)
epl_league_standings_df['epl_season_start_year'] = epl_league_standings_df['epl_season'].map(lambda x: x.split('/')[0])
epl_league_standings_df['epl_season_start_year'] =  pd.to_datetime(epl_league_standings_df['epl_season_start_year'], format='%Y')
epl_league_standings_df['epl_team'].replace('Bournemouth','AFC Bournemouth',inplace=True)
epl_league_standings_df
```

```python
# EPL Team Home-Away Statistics per Season
epl_league_home_away_form_df = pd.read_csv('EPL_Match_Stats.csv')
epl_league_home_away_form_df = epl_league_home_away_form_df[['HomeTeam','AwayTeam','FTR','Season']]
epl_season_filter = ['2006-07','2007-08','2008-09','2009-10','2010-11','2011-12','2012-13','2013-14','2014-15','2015-16','2016-17']
epl_league_home_away_form_df = epl_league_home_away_form_df[epl_league_home_away_form_df['Season'].isin(epl_season_filter)].reset_index(drop=True)
epl_league_home_away_form_df
```

```python
# EPL Team Home Wins per Season
epl_league_home_form_df = epl_league_home_away_form_df[epl_league_home_away_form_df['FTR'] == 'H']
epl_league_home_form_df = epl_league_home_form_df.groupby(['HomeTeam','Season'])['FTR'].count().to_frame().reset_index()
epl_league_home_form_df.columns = ['epl_team','epl_season','home_wins']
epl_league_home_form_df
```

```python
# EPL Team Away Wins per Season 
epl_league_away_form_df = epl_league_home_away_form_df[epl_league_home_away_form_df['FTR'] == 'A']
epl_league_away_form_df = epl_league_away_form_df.groupby(['AwayTeam','Season'])['FTR'].count().to_frame().reset_index()
epl_league_away_form_df.columns = ['epl_team','epl_season','away_wins']
epl_league_away_form_df.loc[-1] = ['Derby','2007-08', 0]
epl_league_away_form_df.loc[-2] = ['Hull','2009-10',0]
epl_league_away_form_df.reset_index(drop=True)
```

```python
# Merging both datasets
epl_league_home_away_form_df= pd.merge(epl_league_home_form_df,epl_league_away_form_df, on=['epl_team','epl_season'])
epl_league_home_away_form_df['epl_season'] = epl_league_home_away_form_df['epl_season'].map(lambda x: x.split('-')[0]+"/20"+x.split('-')[1])
epl_league_home_away_form_df
```

```python
# Removing inconsistencies in EPL team names
epl_league_home_away_form_df['epl_team'].replace(['Tottenham','Man City','Man United','West Brom','West Ham','Leicester','Stoke','Swansea','Hull','Newcastle','QPR','Norwich','Cardiff','Bolton','Blackburn','Wolves','Birmingham','Derby','Charlton','Bournemouth'],['Tottenham Hotspurs','Manchester City','Manchester United','West Bromwich Albion','West Ham United','Leicester City','Stoke City','Swansea City','Hull City','Newcastle United','Queens Park Rangers','Norwich City','Cardiff City','Bolton Wanderers','Blackburn Rovers','Wolverhampton Wanderers','Birmingham City','Derby County','Charlton Athletic','AFC Bournemouth'],inplace=True)
```

```python
epl_league_home_away_form_df['epl_team'].unique()
```

```python
# Merging everything in a Single Dataframe
epl_league_stats = pd.merge(epl_league_home_away_form_df,epl_league_standings_df, on=['epl_team','epl_season']) 
epl_league_stats
```

```python
# Checking if there are NaN values in epl_league_stats
epl_league_stats.isnull().values.any()
```

```python
# EPL Game Statistics Dataset
epl_team_match_stats_df = pd.read_csv('EPL_Team_Wise_Indepth_Data.csv')
epl_team_match_stats_df = epl_team_match_stats_df[['team','att_hd_goal','att_pen_goal','att_freekick_goal','att_ibox_goal','att_obox_goal','goal_fastbreak','clean_sheet','total_pass','total_long_balls','total_through_ball','total_cross','backward_pass','season']]
epl_team_match_stats_df['season'] = epl_team_match_stats_df['season'].map(lambda x: x.split('-')[0]+"/"+x.split('-')[1])
epl_season_filter = ['2006/2007','2007/2008','2008/2009','2009/2010','2010/2011','2011/2012','2012/2013','2013/2014','2014/2015','2015/2016','2016/2017']
epl_team_match_stats_df = epl_team_match_stats_df[epl_team_match_stats_df['season'].isin(epl_season_filter)]
epl_team_match_stats_df.columns = ['epl_team','header_goals','penalty_goals','freekick_goals','insidebox_goals','outsidebox_goals','counter_attack_goals','clean_sheet','total_passes','total_long_balls','total_through_balls','total_crosses','total_backward_pass','epl_season']
epl_team_match_stats_df.fillna(epl_team_match_stats_df.mean(), inplace=True)
epl_team_match_stats_df
```

```python
# Checking if there are NaN values in epl_team_match_stats_df
epl_team_match_stats_df.isnull().values.any()
```

```python
# Removing inconsistencies in EPL team names
epl_league_stats['epl_team'].replace(['Tottenham Hotspurs','Wigan'],['Tottenham Hotspur','Wigan Athletic'],inplace=True)
```

```python
epl_league_stats.to_csv('epl_league_stats.csv')
```

```python
# Also Including the teams who have not played in a particular season  
epl_league_stats['season_position'] = epl_league_stats['season_position'].astype(str)
last_10_years_epl_teams = epl_league_stats['epl_team'].unique()
epl_season_start_year = epl_league_stats['epl_season_start_year'].unique()
epl_league_stats_teams_not_played = pd.DataFrame(columns=epl_league_stats.columns)
for start_year in epl_season_start_year:
    epl_teams = epl_league_stats.loc[epl_league_stats['epl_season_start_year'] == start_year, 'epl_team']
    teams_not_played = list(set(last_10_years_epl_teams)^set(epl_teams))
    epl_league_stats_teams_not_played['epl_team'] = teams_not_played 
    epl_league_stats_teams_not_played['epl_season_start_year'] = start_year
    epl_league_stats_teams_not_played['season_position'] = 'NP'
    epl_league_stats_teams_not_played['epl_season'] = epl_league_stats['epl_season'][epl_league_stats['epl_season_start_year'] == start_year].unique()[0]
    epl_league_stats = epl_league_stats.append(epl_league_stats_teams_not_played)
epl_league_stats    
```

```python
epl_league_stats.to_csv('epl_league_stats_with_NP_Teams.csv')
```
